```python
# importar bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
# GRAFICO COM A MATPLOTLIB - Plotagem de funções matemáticas (integração com Numpy)

# criar uma matriz com a função linspace da numpy (criar uma matriz com base numa qtde de valores e qtde de divsões)
t = np.linspace(0.0, 2.0, 100)

# criar mais duas matrizes com diferentes funções trigonométricas, no caso sin = seno
f1 = np.sin(np.pi * t)
f2= np.sin(2 * np.pi * t)

#definir a figura com base nos dados acima
plt.figure(figsize=(10,8))

plt.plot(t, f1, linewidth=5.0, label='seno de pi * t')
plt.plot(t, f2, linewidth=5.0, label='2 senos de pi * t')

# adicionar legendas, títulos, etc
plt.legend()

plt.title('Senoides')

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5)
```

# GRÁFICO DE VIOLIN

```python
# GRÁFICO COM A PLOTLY - casamento entre histograma e boxplot
# analisar a variação do valor de crédito segundo estado civil

credit = pd.read_excel('/content/clientes_atualizado.xlsx')
credit
```

```python
# definir figura utilizar o submodulo express da plotly definindo os eixos x e y
# conseguimos identificar o outlier que define qual o que está fora do padrão

fig = px.violin(credit, x='estado_civil', y='valor_credito', color='sexo', box=True, title='valor de credito concedido segundo o estado civil dos clientes')

fig.show()
```

# GRÁFICO DE DISPERSÃO SCATTER

```python
# GRÁFICO COM A PLOTLY
# sintese da situação dos clientes no mes de Abril para o ano de 2005 segundo estado civil

# definir a figura incluindo as informações do eixo x e y
fig = px.scatter(credit, x='status_pagto_abr_2005', 
                 y='valor_pagto_abr_2005', 
                 color='estado_civil', 
                 size='valor_credito',
                 title='Sintese da situação dos clientes para o mes de Setembro de 2005 segundo o estado civil')

fig.show()

#integrar o grafico com o git hub

#salvar o grafico no formato interativo
fig.write_html('sintese_clientes_sbr_2005.html')
```
